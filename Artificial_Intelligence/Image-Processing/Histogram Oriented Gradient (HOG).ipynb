```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from skimage.feature import hog
from skimage import data,exposure
```

## Reading image

```python
img = cv2.imread("avengers.jpg")
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## HOG

```python
fd,hog_img = hog(img,orientations=9,pixels_per_cell=(8,8),cells_per_block=(2,2),visualize=True,channel_axis=2)
```

## Rescaling

```python
scale = exposure.rescale_intensity(hog_img, in_range=(0,5))
```

## Plot

```python
figure,(a1,a2)=plt.subplots(1,2, figsize=(12,6), sharex=True, sharey=True)
a1.axis("off")
a1.imshow(img2)
a1.set_title("input image")
a2.axis("off")
a2.imshow(scale)
a2.set_title("output image")
plt.show()
```

```python
cv2.imshow("orginal",img)
cv2.imshow("HOG",scale)
cv2.waitKey(0)
```

```python
import imutils
```

## Initilazion of the HOG person

## Detector

```python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
img2 = cv2.imread("civil-war-avengers.jpg")
cv2.imshow("Orginal",img2)
cv2.waitKey(0)
```

```python
image=imutils.resize(img2,width=min(400,img2.shape[1]))
```

```python
(regions,img3) = hog.detectMultiScale(image,winStride=(2,2),padding=(4,4),scale=1)
```

```python
for(x,y,w,h) in regions:
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,225,0),2)
```

```python
cv2.imshow("Orginal",image)
cv2.waitKey(0)
```
